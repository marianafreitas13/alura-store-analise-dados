### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento_loja1 = loja["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

print("Faturamento Loja 1:", faturamento_loja1)
print("Faturamento Loja 2:", faturamento_loja2)
print("Faturamento Loja 3:", faturamento_loja3)
print("Faturamento Loja 4:", faturamento_loja4)
```

# 2. Vendas por Categoria

```python
print("Categorias Loja 1")
print(loja["Categoria do Produto"].value_counts())

print("\nCategorias Loja 2")
print(loja2["Categoria do Produto"].value_counts())

print("\nCategorias Loja 3")
print(loja3["Categoria do Produto"].value_counts())

print("\nCategorias Loja 4")
print(loja4["Categoria do Produto"].value_counts())
```

# 3. Média de Avaliação das Lojas


```python
avaliacao_loja1 = loja["Avaliação da compra"].mean()
avaliacao_loja2 = loja2["Avaliação da compra"].mean()
avaliacao_loja3 = loja3["Avaliação da compra"].mean()
avaliacao_loja4 = loja4["Avaliação da compra"].mean()

print("Média Loja 1:", avaliacao_loja1)
print("Média Loja 2:", avaliacao_loja2)
print("Média Loja 3:", avaliacao_loja3)
print("Média Loja 4:", avaliacao_loja4)
```

# 4. Produtos Mais e Menos Vendidos

```python
print("Mais vendidos Loja 1")
print(loja["Produto"].value_counts().head())

print("\nMais vendidos Loja 2")
print(loja2["Produto"].value_counts().head())

print("\nMais vendidos Loja 3")
print(loja3["Produto"].value_counts().head())

print("\nMais vendidos Loja 4")
print(loja4["Produto"].value_counts().head())
```

# 5. Frete Médio por Loja

```python
frete_loja1 = loja["Frete"].mean()
frete_loja2 = loja2["Frete"].mean()
frete_loja3 = loja3["Frete"].mean()
frete_loja4 = loja4["Frete"].mean()

print("Frete médio Loja 1:", frete_loja1)
print("Frete médio Loja 2:", frete_loja2)
print("Frete médio Loja 3:", frete_loja3)
print("Frete médio Loja 4:", frete_loja4)
```

# Relatório de Análise de Desempenho das Lojas – Alura Store

## 1. Objetivo da análise

O objetivo desta análise foi avaliar o desempenho das quatro lojas pertencentes ao e-commerce **Alura Store**, com o intuito de auxiliar o Sr. João na tomada de decisão sobre qual unidade vender para investir em um novo negócio.

A análise foi baseada em dados de vendas, avaliações dos clientes e custos de frete, permitindo identificar quais lojas apresentam melhor desempenho e quais possuem menor potencial competitivo dentro da rede.

---

## 2. Métricas analisadas

Para realizar a análise comparativa entre as lojas, foram consideradas as seguintes métricas:

* Faturamento total das lojas
* Categorias de produtos mais vendidas
* Média de avaliação dos clientes
* Produtos mais vendidos e menos vendidos
* Custo médio de frete

Além disso, foram utilizadas visualizações gráficas com a biblioteca **Matplotlib**, permitindo comparar o desempenho entre as lojas de forma mais clara e intuitiva.

---

## 3. Análise do faturamento

A análise do faturamento total permitiu identificar diferenças no desempenho financeiro entre as quatro lojas.

Os resultados foram os seguintes:

* **Loja 1:** R$ 1.534.509,12
* **Loja 2:** R$ 1.488.459,06
* **Loja 3:** R$ 1.464.025,03
* **Loja 4:** R$ 1.384.497,58

Observa-se que a **Loja 1 apresentou o maior faturamento**, indicando forte desempenho comercial e maior geração de receita. Em seguida aparecem a Loja 2 e a Loja 3, que também apresentam resultados consistentes.

Por outro lado, a **Loja 4 apresentou o menor faturamento entre todas**, demonstrando desempenho inferior em comparação às demais unidades.

---

## 4. Categorias de produtos mais vendidas

A análise das categorias de produtos permitiu identificar quais tipos de itens possuem maior demanda entre os clientes.

Essas informações ajudam a compreender o comportamento de compra dos consumidores e podem indicar oportunidades de crescimento em determinadas categorias de produtos.

Além disso, identificar as categorias mais populares pode auxiliar na definição de estratégias de estoque e marketing.

---

## 5. Avaliação dos clientes

A média de avaliação das compras foi utilizada como indicador de **satisfação dos clientes**.

Avaliações mais altas indicam melhor experiência de compra, maior qualidade no atendimento e maior probabilidade de fidelização dos consumidores.

Esse indicador é importante para avaliar a reputação da loja e a qualidade dos serviços oferecidos.

---

## 6. Produtos mais vendidos

A análise dos produtos mais vendidos permitiu identificar os itens com maior popularidade entre os consumidores.

Esses produtos representam oportunidades estratégicas para aumentar as vendas, pois indicam preferências claras do público.

Da mesma forma, a análise dos produtos menos vendidos pode auxiliar na tomada de decisões relacionadas ao gerenciamento de estoque.

---

## 7. Custo médio de frete

O custo médio de frete também foi analisado, pois pode influenciar diretamente a decisão de compra dos clientes.

Fretes mais elevados podem tornar os produtos menos competitivos no mercado, enquanto fretes mais baixos podem aumentar a atratividade da loja.

Portanto, esse fator também deve ser considerado na análise geral de desempenho das lojas.

---

## 8. Comparação geral entre as lojas

A partir da análise das métricas avaliadas — faturamento, avaliações, vendas de produtos e custo de frete — foi possível observar diferenças claras de desempenho entre as lojas da rede Alura Store.

Algumas lojas apresentam maior eficiência comercial e melhor desempenho geral, enquanto outras apresentam resultados mais modestos.

Essas diferenças permitem identificar quais unidades possuem maior potencial de crescimento e quais apresentam menor desempenho dentro da rede.

---

## 9. Recomendação

Com base nos dados analisados, recomenda-se que o **Sr. João considere vender a Loja 4**.

Essa recomendação é baseada principalmente no fato de que a **Loja 4 apresentou o menor faturamento entre todas as unidades analisadas**, com aproximadamente **R$ 1.384.497,58**, indicando menor desempenho comercial em comparação às demais lojas.

Enquanto isso, a **Loja 1 apresentou o maior faturamento da rede**, demonstrando maior capacidade de geração de receita e melhor desempenho de vendas.

Dessa forma, ao vender a Loja 4, o Sr. João poderá concentrar seus investimentos nas lojas com melhor desempenho, aumentando a eficiência da operação e direcionando recursos para as unidades mais lucrativas.

---

## 10. Conclusão

A análise de dados se mostrou uma ferramenta fundamental para apoiar a tomada de decisões estratégicas.

Através da avaliação de métricas de vendas, satisfação dos clientes e custos operacionais, foi possível identificar padrões relevantes e gerar insights importantes sobre o desempenho das lojas da Alura Store.

Com base nas informações analisadas, conclui-se que a **Loja 4 apresenta o menor desempenho financeiro entre as unidades avaliadas**, sendo a principal candidata para venda.

Essa decisão permitirá ao Sr. João otimizar seus recursos e fortalecer as lojas que apresentam maior potencial de crescimento dentro da rede.
